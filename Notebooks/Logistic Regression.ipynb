```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from sklearn.linear_model import LogisticRegression

np.random.seed(0)
```

### Generate Data

```python
x1 = np.random.uniform(0,10,size=(100))
x2 = np.concatenate((np.random.uniform(0,4.5,size=(50)), np.random.uniform(5.5,10,size=(50))),axis=0)

X = np.array([x1, x2]).T
y = X[:,1] > 5
```

```python
X.shape
```

### Plot Data

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

### Creating Classifier

```python
clf = LogisticRegression(random_state=0).fit(X, y)
```

$\theta_0 = -5; \theta_1 = 0; \theta_2 = 1; $

```python
clf.coef_ = np.array([[0,  1.]])
clf.intercept_ = np.array([-5])
```

```python
h = .02
# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
            edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(-0.5, 0.95, 'y=0', fontsize=14,verticalalignment='top', bbox=props)
plt.text(-0.5, 9.95, 'y=1', fontsize=14,verticalalignment='top', bbox=props)
None
```
